```python
import numpy as np
import matplotlib.pyplot as plt

# Matriz de Hilbert 3x3
H = np.array([[1, 1/2, 1/3],
              [1/2, 1/3, 1/4],
              [1/3, 1/4, 1/5]])

# Vector b original
b_original = np.array([1, 1, 1])

# Resolver el sistema original
x_original = np.linalg.solve(H, b_original)

# Calcular el determinante de la matriz de Hilbert
determinante = np.linalg.det(H)

# Calcular la matriz inversa
matriz_inversa = np.linalg.inv(H)

# Calcular el número de condición
numero_condicion = np.linalg.cond(H)

# Introducimos un pequeño cambio en b
b_modificado = np.array([1, 1.001, 1])

# Resolver el sistema con el vector b modificado
x_modificado = np.linalg.solve(H, b_modificado)

# Representación gráfica de las soluciones
fig, ax = plt.subplots()
ax.plot([0, 1, 2], x_original, label="Solución Original", marker='o')
ax.plot([0, 1, 2], x_modificado, label="Solución con b modificado", marker='x')

ax.set_xticks([0, 1, 2])
ax.set_xticklabels(['x1', 'x2', 'x3'])
ax.set_ylabel('Valores de x')
ax.set_title('Comparación de soluciones')
ax.legend()

plt.grid(True)
plt.show()

determinante, matriz_inversa, numero_condicion, x_original, x_modificado

```
